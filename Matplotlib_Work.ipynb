```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as mpt
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
mpt.plot(x,y)
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
mpt.plot(x,y,marker='o')           # marker=symbols on tred line
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
mpt.plot(x,y,marker='o',color='g')
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
mpt.plot(x,y,marker='*',color='g')
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
mpt.plot(x,y,marker='s',ms=25,color='m')    # ms = marker size
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
mpt.plot(x,y,marker='s',ms=25,mec="r",mfc='y',color='m')     # mec=marker edge color & mfc=marker face color
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
mpt.plot(x,y,c='#800000',marker='P',linewidth=10,ms=25,mfc='k')             # c= color number can we access from internet
                                                              # linewidth = size of tred line
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
mpt.plot(x,y,c='#800000',marker='P',linewidth=10,ms=25,mfc='k')
mpt.title("Iron Block Height & Weight")
mpt.xlabel("Height")
mpt.ylabel("Weight")
mpt.grid(axis=('both'),color='y')
```

```python
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
mpt.plot(x,y,'.-.',linewidth=5,c='#800000',marker='P',ms=10,mfc='k')
mpt.title("Iron Block Height & Weight",color='m',size=15)
mpt.xlabel("Height",color='m',size=15)
mpt.ylabel("Weight",color='m',size=15)
mpt.grid(axis=('both'),color='y',linewidth=5)
```

## Matplotlib.pyplot-2

```python
# Subplot

# plot1
mpt.subplot(2,2,1)
x=np.array(range(100,200,20))
y=np.array(range(1000,2000,200))
mpt.plot(x,y)
mpt.title("PLOT-1",color='m')

#mpt.show()
#plot2
mpt.subplot(2,2,2)
a=np.array([102,150,300,450,500])
b=np.array([10,20,50,70,100])
mpt.plot(a,b,'.-.',color='r')
mpt.title("PLOT-2")

#mpt.show()
#plot3
mpt.subplot(2,2,3)
c=np.array([202,130,350,250,700])
d=np.array([100,255,350,401,500])
mpt.plot(c,d,'*',color='g')
mpt.title("PLOT-3")

#mpt.show()

#plot4
mpt.subplot(2,2,4)
e=np.array([100,150,300,400,500])
f=np.array([10,20,350,40,50])
mpt.plot(e,f,'o',color='k')
mpt.title("PLOT-4")

mpt.show()                                    # subplot must be set on first then after array and plot set.
```
