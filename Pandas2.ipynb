```python
# For loop is use to feature Transformations and also kwon as columns transformations
ab=[]

for x in data["POSTED_BY"]:
    if x=="Dealer":
        ab.append(1)
    elif x=="Owner":
        ab.append(2)
    elif x=="Builder":
        ab.append(3)

```
